```python
# Importar la librerÃ­a pandas para manipulaciÃ³n de datos
import pandas as pd

# Cargar el archivo CSV 'videojuegos_online_telemetria.csv' en un DataFrame de pandas.
# Consejo: Siempre es buena prÃ¡ctica verificar el encoding si el archivo CSV no se carga correctamente.
# Por ejemplo: pd.read_csv('archivo.csv', encoding='latin1')
df = pd.read_csv('videojuegos_online_telemetria.csv')

# Mostrar las primeras 5 filas del DataFrame para una primera inspecciÃ³n de los datos.
# Esto ayuda a entender la estructura y el tipo de datos presentes.
# Consejo: DespuÃ©s de cargar los datos, considera usar `df.info()` para ver los tipos de datos y los valores no nulos, y `df.describe()` para estadÃ­sticas descriptivas de columnas numÃ©ricas.
print('Primeras 5 filas del DataFrame:')
print(df.head())
```

```python
# Contar el nÃºmero de valores duplicados en la columna 'PlayerID'.
# 'duplicated()' devuelve una serie booleana indicando si cada fila es un duplicado.
# Por defecto, `keep='first'` marca como True las ocurrencias duplicadas excepto la primera.
# `.sum()` cuenta cuÃ¡ntos valores True hay, es decir, el nÃºmero de duplicados.
num_duplicates = df['PlayerID'].duplicated().sum()

# Mostrar el resultado del conteo de duplicados.
if num_duplicates > 0:
    print(f"Se encontraron {num_duplicates} valores duplicados en la columna 'PlayerID'.")
    print("Las filas duplicadas (manteniendo la primera apariciÃ³n) son:")
    # Mostrar solo las filas que son duplicados (excluyendo la primera ocurrencia).
    # Consejo: Es importante revisar estas filas para entender la naturaleza de los duplicados antes de eliminarlos.
    display(df[df['PlayerID'].duplicated(keep='first')])
else:
    print("No se encontraron valores duplicados en la columna 'PlayerID'.")
```

```python
# Mostrar todas las filas que tienen valores duplicados en la columna 'PlayerID'.
# `keep=False` marca como True todas las ocurrencias de los duplicados, incluyendo la primera.
# Esto permite ver el grupo completo de filas que comparten un 'PlayerID' duplicado.
duplicated_rows = df[df['PlayerID'].duplicated(keep=False)]

# Verificar si se encontraron filas duplicadas y mostrarlas.
if not duplicated_rows.empty:
    print("Todas las filas con 'PlayerID' duplicados son:")
    # Consejo: Analizar estas filas es crucial para decidir cÃ³mo manejar los duplicados (ej. Â¿cuÃ¡l es la fila 'correcta' si hay inconsistencias?).
    display(duplicated_rows)
else:
    print("No se encontraron valores duplicados en la columna 'PlayerID'.")
```

```python
# Eliminar filas duplicadas basÃ¡ndose Ãºnicamente en la columna 'PlayerID'.
# `subset=['PlayerID']` especifica que la bÃºsqueda de duplicados se haga solo en esta columna.
# `keep='first'` asegura que la primera ocurrencia de cada 'PlayerID' se mantenga y las siguientes se eliminen.
df_cleaned = df.drop_duplicates(subset=['PlayerID'], keep='first')

# Mostrar el nÃºmero de filas antes y despuÃ©s de la eliminaciÃ³n para verificar la operaciÃ³n.
print(f"Filas antes de eliminar duplicados: {df.shape[0]}")
print(f"Filas despuÃ©s de eliminar duplicados: {df_cleaned.shape[0]}")

# Actualizar el DataFrame original `df` con el DataFrame limpio `df_cleaned`.
# Se utiliza `.copy()` para asegurar que `df` sea una copia independiente y evitar `SettingWithCopyWarning` en futuras operaciones.
df = df_cleaned.copy()

# Verificar si aÃºn existen duplicados en 'PlayerID' para confirmar que la eliminaciÃ³n fue exitosa.
num_remaining_duplicates = df['PlayerID'].duplicated().sum()
if num_remaining_duplicates == 0:
    print("\nSe han eliminado todos los duplicados en la columna 'PlayerID'.")
else:
    print(f"\nTodavÃ­a quedan {num_remaining_duplicates} duplicados en 'PlayerID'.")
# Consejo: Siempre verifica el Ã©xito de la operaciÃ³n de limpieza de datos.
```

### Manejo de valores faltantes en la columna 'Age'

Vamos a calcular la mediana de la columna 'Age' y usarla para rellenar los valores faltantes (NaN) en esta columna.

```python
# Calcular la mediana de la columna 'Age'. La mediana es robusta a valores atÃ­picos (outliers).
median_age = df['Age'].median()
print(f"Mediana de la columna 'Age': {median_age}")

# Rellenar los valores faltantes (NaN) en la columna 'Age' con la mediana calculada.
# `df.loc[:, 'Age']` se usa para asegurar la modificaciÃ³n en el DataFrame original y evitar advertencias.
df.loc[:, 'Age'] = df['Age'].fillna(median_age)

# Verificar si quedan valores NaN en la columna 'Age' despuÃ©s de la imputaciÃ³n.
print(f"\nNÃºmero de valores NaN en la columna 'Age' despuÃ©s de rellenar: {df['Age'].isnull().sum()}")

# Mostrar las primeras filas del DataFrame para verificar visualmente el cambio en la columna 'Age'.
print("\nPrimeras 5 filas del DataFrame con 'Age' rellenado:")
print(df.head())

# Mostrar la informaciÃ³n del DataFrame (tipos de datos, conteo de no nulos) para confirmar el estado de 'Age'.
print("\nInformaciÃ³n del DataFrame despuÃ©s de rellenar 'Age':")
# Consejo: `df.info()` es una excelente herramienta para verificar rÃ¡pidamente si las columnas tienen valores nulos y sus tipos de datos.
df.info()
```

```python
# Mostrar los valores distintos en la columna 'UsesVoiceChat'.
# Esta columna es binaria (0 o 1).
print("Valores distintos en 'UsesVoiceChat':")
print(df['UsesVoiceChat'].unique())

# Contar el nÃºmero de valores NaN (faltantes) en la columna 'UsesVoiceChat'.
num_nan_uses_voice_chat = df['UsesVoiceChat'].isnull().sum()
print(f"\nNÃºmero de valores NaN (faltantes) en la columna 'UsesVoiceChat': {num_nan_uses_voice_chat}")

# Si hay valores NaN, se procede a rellenarlos de manera probabilÃ­stica.
if num_nan_uses_voice_chat > 0:
    print("Estos valores 'nan' representan datos faltantes en la columna 'UsesVoiceChat'.")

    # Calcular la proporciÃ³n de 0s y 1s en los datos existentes (no nulos).
    # `normalize=True` devuelve las frecuencias relativas.
    value_counts = df['UsesVoiceChat'].dropna().value_counts(normalize=True)
    prob_0 = value_counts.get(0, 0) # Probabilidad de 0 (no usa chat de voz)
    prob_1 = value_counts.get(1, 0) # Probabilidad de 1 (usa chat de voz)

    print(f"\nProporciÃ³n de 0s (no usa chat de voz): {prob_0:.2f}")
    print(f"ProporciÃ³n de 1s (usa chat de voz): {prob_1:.2f}")

    # Rellenar los valores NaN probabilÃ­sticamente, manteniendo la distribuciÃ³n original.
    # `np.random.choice` selecciona 0 o 1 basÃ¡ndose en las proporciones calculadas.
    import numpy as np
    fill_values = np.random.choice([0, 1], size=num_nan_uses_voice_chat, p=[prob_0, prob_1])

    # Obtener los Ã­ndices de los valores NaN para asignar los valores rellenados.
    nan_indices = df[df['UsesVoiceChat'].isnull()].index

    # Asignar los valores rellenados a los NaN en el DataFrame.
    df.loc[nan_indices, 'UsesVoiceChat'] = fill_values

    print("\nValores distintos en 'UsesVoiceChat' DESPUÃS de rellenar NaN:")
    print(df['UsesVoiceChat'].unique())

    # Verificar nuevamente si quedan valores NaN despuÃ©s de la imputaciÃ³n.
    num_nan_uses_voice_chat_after = df['UsesVoiceChat'].isnull().sum()
    print(f"\nNÃºmero de valores NaN (faltantes) en la columna 'UsesVoiceChat' DESPUÃS de rellenar: {num_nan_uses_voice_chat_after}")

    # Confirmar si todos los valores NaN han sido rellenados.
    if num_nan_uses_voice_chat_after == 0:
        print("Todos los valores NaN en 'UsesVoiceChat' han sido rellenados con Ã©xito.")
# Si no hay NaN inicialmente, informar de ello.
else:
    print("No se encontraron valores NaN en la columna 'UsesVoiceChat'.")

# Mostrar el conteo de valores por cada 'UsesVoiceChat', incluyendo NaN (se espera 0 despuÃ©s de rellenar).
print("\nConteo de valores por cada 'UsesVoiceChat' (incluyendo NaN despuÃ©s de rellenar si aplica):")
print(df['UsesVoiceChat'].value_counts(dropna=False))
```
